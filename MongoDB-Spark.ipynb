```
from pyspark.sql import SparkSession

#https://stackoverflow.com/questions/69470250/how-to-connect-to-mongodb-atlas-from-databricks-cluster-using-pyspark
#Install org.mongodb.spark:mongo-spark-connector_2.12:3.0.2 maven library to your cluster if using scala2.12 in cluster config

# Configure your MongoDB Atlas connection settings
#mongo_uri = "mongodb+srv://mongodb:mongodb@kafka-mongo.cgjosxp.mongodb.net?retryWrites=true&w=majority"
mongo_uri_case = "mongodb+srv://mongodb:mongodb@kafka-mongo.cgjosxp.mongodb.net/kafka_mongo_covid_ds.case_1?retryWrites=true&w=majority"
mongo_uri_region = "mongodb+srv://mongodb:mongodb@kafka-mongo.cgjosxp.mongodb.net/kafka_mongo_covid_ds.region_1?retryWrites=true&w=majority"
mongo_uri_time_province = "mongodb+srv://mongodb:mongodb@kafka-mongo.cgjosxp.mongodb.net/kafka_mongo_covid_ds.time_province_1?retryWrites=true&w=majority"


# Create a SparkSession in DataBricks
spark = SparkSession.builder \
    .appName("MongoDB to PySpark DataFrame") \
    .getOrCreate()
#     .config("spark.mongodb.input.uri", mongo_uri) \
#     .config("spark.mongodb.input.readPreference.name", "secondaryPreferred") \
    

# Read data from MongoDB into a DataFrame
df_case = spark.read.format("mongo").option("uri", mongo_uri_case).load()
df_region = spark.read.format("mongo").option("uri", mongo_uri_region).load()
df_time_province = spark.read.format("mongo").option("uri", mongo_uri_time_province).load()

# Show the DataFrame
df_case.show(5)
display(df_case)
df_region.show(5)
display(df_region)
df_time_province.show(5)
display(df_time_province)
```

```
record_count_case = df_case.count()
print("Number of records in 'case' collection:", record_count_case)

record_count_region = df_region.count()
print("Number of records in 'region' collection:", record_count_region)

record_count_time_province = df_time_province.count()
print("Number of records in 'time_province' collection:", record_count_time_province)
```

```
df_case.describe().show()
df_region.describe().show()
df_time_province.describe().show()
```

```
df_case = df_case.drop('_id')
df_region = df_region.drop('_id')
df_time_province = df_time_province.drop('_id')
```

```
df_case = df_case.dropDuplicates()
df_region = df_region.dropDuplicates()
df_time_province = df_time_province.dropDuplicates()

print("Number of records in 'case' Data Frame after removing Duplicates:", df_case.count())
print("Number of records in 'region' Data Frame after removing Duplicates:", df_region.count())
print("Number of records in 'time_province' Data Frame after removing Duplicates:", df_time_province.count())
```

```
df_case = df_case.withColumnRenamed('infection_case', 'infection_case_grp_name')

df_case.limit(5).show()
```

```
subset_df = df_case.select('city', 'confirmed')
#Select all Columns
#subset_df = df_case.select('*')
subset_df.show()
display(subset_df)
```

```
# Assuming you already have a DataFrame named 'df'
df_case = df_case.fillna("null")

# Show the DataFrame with null values filled
df_case.show()
display(df_case)
```

```
df_case = df_case.na.replace("-", "null")
df_case.show()
display(df_case)
```

```
#Below Command Drops Rows from City Column alone if there is NA
df_case = df_case.dropna(subset=['city'])
df_case.show()
display(df_case)
```

```
df_case = df_case.filter((df_case['city'] != "null") & (df_case['latitude'] != "null") & (df_case['longitude'] != "null"))
df_case.show()
display(df_case)
```

```
#Find out all the different infection_case in Daegu Province with more than 10 confirmed cases.
df_infection_daegu = df_case.filter((df_case['province'] == "Daegu") & (df_case['confirmed'] > 10)).select("infection_case_grp_name")

infection_daegu_count = df_infection_daegu.distinct().count()
print(infection_daegu_count)
df_infection_daegu.show()
display(df_infection_daegu)

#distinct_count = df.select(col(column)).distinct().count()
```

```
#Sort the number of confirmed cases ascending
df_case_asc=df_case.orderBy("confirmed")
df_case_asc.show()
display(df_case_asc)
```

```
#Sort the number of confirmed cases descending

from pyspark.sql.functions import col
from pyspark.sql.functions import desc

df_case_desc = df_case.orderBy(desc("confirmed"))
#OR
#df_case_desc=df_case.orderBy(col("confirmed").desc())
df_case_desc.show()
display(df_case_desc)
```

```
#Cast that data type 
df_case = df_case.withColumn("latitude", df_case["latitude"].cast("double"))
df_case = df_case.withColumn("longitude", df_case["longitude"].cast("double"))
df_case.describe()

df_case.show()
display(df_case)
```

```
#Use group by on top of province and city column and agg it with sum of confirmed cases

from pyspark.sql import functions as func
df_case_agg=df_case.groupBy(["province","city"]).agg(func.sum("confirmed"))
df_case_agg.show()
display(df_case_agg)
```

```
#Use different different join methods
df_result = df_case.join(df_region, ['province','city'],how='left')
df_result.show()
display(df_result)
```

```
#If case is less than 50 return low, else return high. Convert into a UDF Function and mention the return type of function.
#casehighlow()

from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

@udf(returnType=StringType())
def casehighlow(case_value):
    if case_value < 50:
        return 'low'
    else:
        return 'high'

df_case_udf = df_case.withColumn("case_high_low", casehighlow("confirmed"))
df_case_udf.show()
display(df_case_udf)
```
